```python
gstasesor=bd_gstasesor[bd_gstasesor['RESPUESTA_OBTENIDA'].notnull()].copy()
```

```python
gstasesor['RESPUESTA_OBTENIDA'] = gstasesor['RESPUESTA_OBTENIDA'].astype('category')
ordenCalltype=['ACUERDO DE PAGO','ACUERDO DE CONVENIO','LOCALIZADO SIN ACUERDO',\
               'YA PAGO','RECLAMO','RENUENTE','RECLAMO VENDIO VEHICULO','VOLVER A LLAMAR',\
               'INFRACTOR CUELGA LLAMADA','ACTUALIZACION DATOS','MENSAJE CON TERCERO',\
               'FALLECIDO / OTROS','AQUI NO VIVE / TRABAJA','BUZON DE MENSAJES',\
               'TELEFONO OCUPADO','TELEFONO DANADO','NO CONTESTA']
gstasesor['RESPUESTA_OBTENIDA'] = gstasesor['RESPUESTA_OBTENIDA'].cat.set_categories(ordenCalltype)
```

```python
gstasesor = gstasesor.sort_values(by=['RESPUESTA_OBTENIDA'],ascending=[True])
```
